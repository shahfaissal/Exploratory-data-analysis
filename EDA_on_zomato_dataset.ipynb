```python
df_country = pd.read_csv("/content/drive/MyDrive/Country-Code.csv",encoding = 'unicode_escape')
```

```python
df_country.head()
```

```python
#combining two dataframes
final_df = pd.merge(df,df_country,on = 'Country Code',how='left')
```

```python
final_df.head(2)
```

```python
country_names= final_df['Country'].value_counts().index
```

```python
country_values = final_df['Country'].value_counts().values
```

```python
#pie chart - top 3 countries which use zomato
plt.pie(country_values[:3],labels=country_names[:3],autopct="%1.2f%%")
```

**observation** :

zomato records maximum transcations or records in india followed by USA and then United kingdoms

```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns= {0:'rating count'})
```

```python
ratings 
```

```python
plt.rcParams['figure.figsize']=(15,8)
sns.barplot(x='Aggregate rating',y='rating count',data=ratings)
```

```python
sns.barplot(x='Aggregate rating',y='rating count',hue='Rating color',data=ratings,palette=['white','red','orange','yellow','green','green'])
```

**observation:**

1.Not rated count is very high

2.maximum number of ratings are between 2.5 to 3.4


```python
#count plot
sns.countplot(x='Rating color',data=ratings,palette=['blue','red','orange','yellow','green','green'])
```

```python
#find the countries name that has given 0 rating 
country_zero=final_df[final_df['Aggregate rating']==0]
```

```python
country_zero[['Country','Aggregate rating']].value_counts().reset_index()
```

**observations:**

1.maximum number of 0 ratings are from indian customers

```python
#find out which currency is used by which country?
final_df.groupby(['Country','Currency']).size().reset_index()
```

```python
#which countries do have online delivery option
delivery_countries = final_df[final_df['Has Online delivery']=='Yes']
```

```python
delivery_countries[['Country','Has Online delivery']].value_counts().reset_index()
```

**observations:**

1.Online deliveries is availble in few parts of India and Uae

```python
#create a pie chart for city distribution
city_names= final_df['City'].value_counts().index
city_values = final_df['City'].value_counts().values
```

```python
#pie chart for order distribution in cities
plt.pie(city_values[:5],labels=city_names[:5],autopct="%1.2f%%")
plt.title("food order distribution chart across various cities of zomato users ")
```
